The dataset discussed was downloaded from `https://archive.ics.uci.edu/ml/machine-learning-databases/iris/`

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Upscale the plots
plt.rcParams['figure.dpi'] = 200

iris = pd.read_csv("Iris_Data_Set/iris.data")
```

```python
# How many data points and features?
print(type(iris)) # iris is a Pandas DataFrame
print(iris.shape) # 150 Data points, and 5 features
```

```python
# What are the column names for our dataset?
print(iris.columns)
```

```python
# How many data points for each class are present?
# The class label is 'species'
iris['species'].value_counts()
```

```python
# 2D scatter plot:
iris.plot(kind='scatter', x='sepal_length', y='sepal_width'); plt.show()
```

```python
# 2D scatter plot with color coding for each class.
sns.set_style('whitegrid')
sns.FacetGrid(iris, hue="species", height=4) \
    .map(plt.scatter, 'sepal_length', 'sepal_width') \
    .add_legend()
plt.show()
```
